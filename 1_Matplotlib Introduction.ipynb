```python
import  numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
x
```

```python
y
```

Functional Method

```python
plt.plot(x,y,'g')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

**Subplot**

```python
plt.subplot(2,2,1)   
plt.plot(x,y,'r--')


plt.subplot(2,2,4)
plt.plot(y,x,'b*-')
```

Object Oriented (OO) Method

OO method more efficeint and controlable than Function method

```python
fig=plt.figure()

axes=fig.add_axes([1,1,2,2])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig=plt.figure()

axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.4,0.2,0.4,0.3])
```

```python
fig=plt.figure()

axes1= fig.add_axes([0.1,0.1,0.9,0.9])
axes2= fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_title('Larger Plot')

axes2.plot(y,x,'g')
axes2.set_title('Smaler plot')
```
