# Sub Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y= x**2
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
fig,axes=plt.subplots()
axes.plot(x,y)
```

```python
axes=fig.add_axes([1,1,8,8])
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for ax in axes:
    ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'g')           
axes[0].set_title('First Figure')
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][1].plot(x,y,'r')
```

# Figure Size and DPI (Dots Per Inch)

```python
fig=plt.figure(figsize=(3,2))

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig=plt.figure(figsize=(8,2))

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes= plt.subplots(nrows=2,ncols=1,figsize=(8,4))

axes[0].plot(x,y)
axes[1].plot(y,x,'g')

plt.tight_layout()
```

```python
fig.savefig('first.pdf',dpi=300)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared')  # Figure legend eklemek icin plot icinde "label" tanimlamaliyiz
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=0)    # loc=0 legend icin best yeri secer
                    # loc(0.1,0,1) gibi tuple icinde nokta da secebiliriz 
```
